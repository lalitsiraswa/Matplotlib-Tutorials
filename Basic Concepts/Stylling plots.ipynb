```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 20)
y = x * x
z = x + y
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1, 1])
axes1.plot(y, x, color = "black")

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (7, 7))
axes[0][0].plot(y, x, color = "black")
axes[0][1].plot(y, z, color = "red")
axes[1][0].plot(x, y, color = "blue")
axes[1][1].plot(z, x, color = "green")
plt.tight_layout()
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1, 0.1, 1.3, 1])
x1.plot(x, x, color = "y")
x1.set_xlabel("X - axis")
x1.set_ylabel("Y - axis")
x1.set_title("A Graph")

x2 = fig.add_axes([0.2, 0.5, 0.5, 0.5])
x2.plot(y, x, color = "b")
x2.set_xlabel("X - label")
x2.set_ylabel("Y - label")
x2.set_title("B Graph")

x3 = fig.add_axes([0.8, 0.5, 0.5, 0.5])
x3.plot(x, z, color = "g")
x3.set_xlabel("X - label")
x3.set_ylabel("Y - label")
x3.set_title("C Graph")

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (7, 6))
axes[0][0].plot(x, y, color = "black", linewidth = 5)
axes[0][1].plot(y, x, color = "green", linewidth = 6)
axes[1][0].plot(y, z, color = "yellow", linewidth = 7)
axes[1][1].plot(z, x, color = "brown", linewidth = 5)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (7, 6))
axes[0][0].plot(x, y, color = "black", lw = 5)
axes[0][1].plot(y, x, color = "green", lw = 6)
axes[1][0].plot(y, z, color = "yellow", lw = 7)
axes[1][1].plot(z, x, color = "brown", lw = 5)

plt.tight_layout()
```
