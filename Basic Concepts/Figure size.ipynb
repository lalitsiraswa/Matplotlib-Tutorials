```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 20)
y = x * x
z = x + y
```

```python
fig = plt.figure(figsize = (4, 5))
x1 = fig.add_axes([0.1, 0.1, 1, 1])
x2 = fig.add_axes([0.3, 0.3, 0.5, 0.5])
plt.tight_layout()
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1, 0.1, 1, 1])
x2 = fig.add_axes([0.3, 0.3, 0.5, 0.5])
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 1, figsize = (4, 5))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```
