```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 21, 20)
y = x * x
z = x + y
```

```python
print(x)
print(y)
print(z)
```

```python
plt.subplot(3, 2, 1)
plt.plot(x, y)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("A")

plt.subplot(3, 2, 2)
plt.plot(x, z)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("B")

plt.subplot(3, 2, 3)
plt.plot(y, z)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("C")

plt.subplot(3, 2, 4)
plt.plot(x, x)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("D")

plt.subplot(3, 2, 5)
plt.plot(y, y)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("E")

plt.subplot(3, 2, 6)
plt.plot(z, z)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("F")

plt.tight_layout()
```
