```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = x * x
z = x + y
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.set_xlabel("X - axis")
axes.set_ylabel("Y - axis")
axes.set_title("First Graph")
axes.plot(z, x)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1, 0.1, 1.3, 1])
x1.plot(x, y)
x1.set_xlabel("X - axis")
x1.set_ylabel("Y - axis")
x1.set_title("A Graph")

x2 = fig.add_axes([0.2, 0.5, 0.5, 0.5])
x2.plot(y, x)
x2.set_xlabel("X -axis")
x2.set_ylabel("Y - axis")
x2.set_title("B Graph")

x3 = fig.add_axes([0.8, 0.5, 0.5, 0.5])
x3.plot(z, x)
x3.set_xlabel("X - axis")
x3.set_ylabel("Y - axis")
x3.set_title("C Graph")

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0][0].plot(x, y)
axes[0][0].set_xlabel("X - axis")
axes[0][0].set_ylabel("Y - axis")
axes[0][0].set_title("A Graph")

axes[0][1].plot(y, z)
axes[0][1].set_xlabel("X - axis")
axes[0][1].set_ylabel("Y - axis")
axes[0][1].set_title("B Graph")

axes[1, 0].plot(z, x)
axes[1, 0].set_xlabel("X - axis")
axes[1][0].set_ylabel("Y - axis")
axes[(1, 0)].set_title("C Graph")

axes[1, 1].plot(x, x)
axes[1, 1].set_xlabel("X - axis")
axes[1][1].set_ylabel("Y - axis")
axes[1][1].set_title("D Graph")

plt.tight_layout()
```
