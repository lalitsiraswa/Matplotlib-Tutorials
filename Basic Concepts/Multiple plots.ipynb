```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 10)
y = x * x
z = x + y
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(x, y)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1, 0.1, 1, 1])
x2 = fig.add_axes([0.1, 0.1, 1, 1])
x1.plot(x, y)
x2.plot(y, x)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1, 0.1, 1.1, 1.3])
x2 = fig.add_axes([0.2, 0.8, 0.4, 0.4])
x3 = fig.add_axes([0.7, 0.8, 0.4, 0.4])
x1.plot(x, y)
x2.plot(y, z)
x3.plot(z, x)
```
