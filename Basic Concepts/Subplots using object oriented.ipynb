```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1, 10, 20)
y = x * x
z = x + y
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1, 0.1, 1, 1])
x2 = fig.add_axes([0.4, 0.4, 0.5, 0.5])
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y, y, x)
axes[1].plot(y, z, x, y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0][0].plot(x, y)
axes[0][1].plot(y, x)
axes[1, 0].plot(y, z)
axes[1, 1].plot(z, x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, z)
axes[(0, 1)].plot(z, y)
axes[1][0].plot(y, x)
axes[1][1].plot(y, x)
plt.tight_layout()
```
