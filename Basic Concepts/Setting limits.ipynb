```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 20)
y = x * x
z = x + y
```

```python
plt.plot(x, y, linewidth = "7", color = "black", marker = "o", markersize = "12", markerfacecolor = "blue", markeredgecolor = "black", markeredgewidth = "3")
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(y, x, color = "blue", linewidth = "7", marker = "s", markerfacecolor = "black", markersize = "12", markeredgecolor = "red")

plt.tight_layout()
```

```python
fig = plt.figure()

x1 = fig.add_axes([0.1, 0.1, 1, 1])
x1.plot(x, y, marker = "o")

plt.tight_layout()
```

```python
fig = plt.figure()

x1 = fig.add_axes([0.1, 0.1, 1, 1])
x1.plot(x, y, marker = "o")

x1.set_xlim([0, 5])

plt.tight_layout()
```
