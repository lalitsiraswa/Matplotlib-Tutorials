```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = x * x
z = x + y
```

```python
plt.plot(y, x, color = "black", linewidth = 2, marker = "o")
```

```python
plt.plot(x, y, marker = "s")
```

```python
plt.plot(z, x, marker = "x")
```

```python
plt.plot(x, y, marker = "D")
```

```python
plt.plot(z, x, marker = "p", markersize = 10)
```

```python
plt.plot(x, y, marker = "d", markersize = 20)
```

```python
plt.plot(y, x, marker = "o", markersize = 10, markerfacecolor = "black", markeredgecolor = "red")
```

```python
plt.plot(z, x, color = "black", linewidth = "1", marker = "s", markerfacecolor = "red", markeredgecolor = "green", markersize = 12)
```

```python
plt.plot(y, x, color = "brown", marker = "h", markersize = "10", markerfacecolor = "black", markeredgecolor = "blue", linewidth = "4")
```

```python
plt.plot(z, x, linewidth = "5", marker = "D", markerfacecolor = "black", markersize = "8", markeredgecolor = "blue", markeredgewidth = "2", color = "black")
```
