```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("seaborn")

data = pd.read_csv("files/Subplots_data.csv")
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

plt.plot(ages, py_salaries, label = "Python Devs", linewidth = 5)
plt.plot(ages, js_salaries, label = "JavaScript Devs", linewidth = 5)
plt.plot(ages, dev_salaries, label = "All Devs", linewidth = 5)

plt.title("Median Salary (USB) By Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.legend()

plt.tight_layout()
```
