```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# plt.style.use("fivethirtyeight")
data = pd.read_csv("files/Filling Area on Line Plots.csv")
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
plt.plot(ages, dev_salaries, label = "All Devs", color = "red")
plt.plot(ages, py_salaries, label = "Python Devs", color = "black")
# plt.plot(ages, js_salaries, label = "JavaScript Devs", color = "green")

plt.title("Median Salarty (USD) By Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")

plt.legend()
plt.tight_layout()
```
