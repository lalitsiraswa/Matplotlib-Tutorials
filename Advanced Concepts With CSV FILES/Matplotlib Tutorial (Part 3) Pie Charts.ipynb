```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
slices = [175, 125, 80, 45, 60]
labels = ["1st label", "2nd lebel", "3rd label", "4th label", "5th label"]
colors = ["black", "red", "green", "brown", "blue"]
plt.pie(slices, labels=labels, colors = colors, wedgeprops = {'edgecolor' : "white"})

plt.title("My Awesome Pie Chart")

plt.tight_layout()
```
