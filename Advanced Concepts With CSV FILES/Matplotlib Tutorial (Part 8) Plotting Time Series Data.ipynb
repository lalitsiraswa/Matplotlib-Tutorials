```python
plt.style.use("seaborn")

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle = "-", markersize = 10, marker = "o")

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter("%b %d %Y")

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
```

```python
data = pd.read_csv("files/Plotting Time Series Data.csv")
price_date = data["Date"]
price_close = data["Close"]

plt.plot_date(price_date, price_close, linestyle = "-", markersize = 10)

plt.gcf().autofmt_xdate()

plt.title("Bitcoin Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.tight_layout()
```

```python
plt.style.use("seaborn")
data = pd.read_csv("files/Plotting Time Series Data.csv")

data["Date"] = pd.to_datetime(data["Date"])
data.sort_values("Date", inplace = True)

price_date = data["Date"]
price_close = data["Close"]

plt.plot_date(price_date, price_close, linestyle = "-", markersize = 10)

plt.gcf().autofmt_xdate()

plt.title("Bitcoin Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.tight_layout()

```
