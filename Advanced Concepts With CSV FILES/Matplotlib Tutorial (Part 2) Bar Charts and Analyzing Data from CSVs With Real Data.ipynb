```python
with open("files/data.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row["LanguagesWorkedWith"].split(";"))
print(language_counter)
```

```python
# language_counter
```

```python
print(language_counter.most_common(15))
```

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
print(languages, "\n")
print(popularity)
```

```python
plt.bar(languages, popularity)

plt.xlabel("Programming Languages")
plt.ylabel("Number Of People Who Use")
plt.title("Most Popular Languages")

plt.tight_layout()
```

```python
# Horizontal Bar Chart
plt.barh(languages, popularity)

# plt.ylabel("Programming Languages")
plt.xlabel("Number Of People Who Use")
plt.title("Most Popular Languages")

plt.tight_layout()
```

```python
# Horizontal Bar Chart
languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)

# plt.ylabel("Programming Languages")
plt.xlabel("Number Of People Who Use")
plt.title("Most Popular Languages")

plt.tight_layout()
```

```python
import csv
from collections import Counter
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("files/data.csv")
data.head()
```

```python
ids = data["Responder_id"]
lang_responces = data["LanguagesWorkedWith"]
print(ids.head(), "\n")
print(lang_responces.head())
```

```python
language_counter = Counter()
lang_responces.loc[0].split(";")
```

```python
for response in lang_responces:
    language_counter.update(response.split(";"))
```

```python
language_counter.most_common(15)
```

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
print(languages)
print(popularity)
```

```python
plt.bar(languages, popularity)
```

```python
languages.reverse()
popularity.reverse()
```

```python
plt.barh(languages, popularity, color = "black")

plt.xlabel("Number Of People Who Use")
plt.title("Most Popular Languages")

plt.tight_layout()
```
