```python
plt.style.use('seaborn')

data = pd.read_csv("files/Scatter Plots.csv")
view_count = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

plt.scatter(view_count, likes, edgecolor = "black", s = 100, linewidth = 2, alpha = 0.75, c = ratio, cmap="summer")

plt.xscale("log")
plt.yscale("log")

cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")

plt.title("Trending YouTube Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")

plt.tight_layout()
```
